Bitwise Operators
Bitwise operators are used to compare (binary) numbers:
&, |-or , ^-=XOR, ~ = not inverts all the bits, <<= zero fill left shift,>>=signed right shift

mostly use in robtic.

```python
#Bitwise and
print(2 & 3)

#bitwise or
print(2 | 3)

#bitwise xor
print(2^3)

#bitwise not
print(~3)

#bitwise right shift
print(4>>2)
#bitwise left shift.
print(4<<2)
```
