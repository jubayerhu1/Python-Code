```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Volumes/study/A_data-science/Python-Code/Feature Engineering /.wine_data.csv', header = None, usecols = [0,1,2])
```

```python
df.columns = ['class label', 'Alcohol', 'Malic acid']
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict = {1: 'red', 3: 'green', 2:'blue'}

sns.scatterplot( 
    x = df['Alcohol'],
    y=df['Malic acid'],
    hue =df['class label'], palette = color_dict,
   
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df.drop('class label', axis = 1),
                                    df['class label'], test_size = 0.3, random_state =0)

```

```python
X_train.shape, X_test.shape
```

```python
Y_train.shape , Y_test.shape
```

```python
# Min Max Scaler

from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()

# fit the scaler to train set, it will learn the parameters
Scaler.fit(X_train)

# transform train and test sets
X_train_scaled = Scaler.transform(X_train)
X_test_scaled = Scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# Convert data frame 

X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize = (12,5))

ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c = Y_train)
ax1.set_title("Before Scaling ")

ax2.scatter(X_train_scaled["Alcohol"], X_train_scaled['Malic acid'], c =Y_train)
ax2.set_title("After Scaling ")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize =(12,5))

# before scaling 
ax1.set_title("Alcohol Distribution Before Scaling ")
sns.kdeplot(X_train['Alcohol'], ax = ax1)

# After scaling 
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax =ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize =(12,5))

# before scaling 
ax1.set_title('Malic acid Distribution before scaling')
sns.kdeplot(X_train['Malic acid'], ax = ax1)

# After scaling 
ax2.set_title("Malic acid distribution After Standar scaling ")
sns.kdeplot(X_train_scaled['Malic acid'], ax =ax2)
plt.show()
```
