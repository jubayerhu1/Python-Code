```python
from sklearn .impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.isnull().sum()
```

```python
si = SimpleImputer()
```

```python
df.head()
```

```python
# adding simple imputer to fever column
X_train_fever = si.fit_transform(X_train[['fever']])

# also the test data
X_test_fever = si.fit_transform(X_test[['fever']])

X_train_fever.shape
```

```python
X_train_fever
```

```python
# Ordinalencoding -> cough
oe = OrdinalEncoder(categories = [['Mild', 'Strong']])# apply the mild and strong 
X_train_cough = oe.fit_transform(X_train[['cough']])

# Also the test data 
X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```python
X_train_cough
```

```python
# OneHotEncoding - > gender, city

ohe = OneHotEncoder(drop='first',sparse_output=False)

X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
X_train_gender_city
```

```python
# Extracting Age 
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# Also the test dat 
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
# Concatenate 
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city),axis=1) # axis=1 column wise

X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city),axis=1)

X_train_transformed.shape
```

```python
# with column Transformer

from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer( transformers= [
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False, drop='first'),['gender','city'])
],remainder='passthrough')

# has_covid colum we don't remove or edit 
```
