# A control statement
## In Python, a conditional statement allows the program to execute different blocks of code based on whether a specified condition evaluates to True or False. This enables decision-making and control flow within a program. The primary keywords used for conditional statements in Python are if, elif, and else.

# if, elif and else 

### In Python, if, elif, and else statements are used for conditional execution,
### allowing a program to make decisions and execute different blocks of code based on whether certain conditions are true or false

```python
# if statement : 
## The if statement executes a block of code only if its condition is True.

x = 10 
if x > 5 :
    print( " x is greater than 5")
```

```python
# #2. if-else statement:
# The if-else statement provides an alternative block of code to execute if the if condition is False.
# Python
x = 3
if x > 5 :
    print(" x is greater than 5")
else:
    print(" x is not greater than 5")
 
```

```python
# 3. if-elif-else statement:
# The if-elif-else statement allows for checking multiple conditions sequentially. 
# If the if condition is False, the program checks the elif conditions in order.
# If none of the if or elif conditions are True, the else block is executed.

grade = int(input("enter your number"))
if grade >= 90:
    print(" Excellent !")
elif grade >=70:
    print("Good job !")
else:
    print("keep practicing --")
```

Python Nested if Statements
It is possible to include an if statement inside another if statement. For example,
number = 5

# outer if statement
if number >= 0:
    # inner if statement
    if number == 0:
      print('Number is 0')
    
    # inner else statement
    else:
        print('Number is positive')

# outer else statement
else:
    print('Number is negative')

```python
number = int(input("Enter your number :"))

if number >= 0:
    #inner if statement;
    if number == 0:
        print("number is 0",number)
    # inner else statement 
    else:
        print(" Number is positive ",number)
# outer else statement
else:
    print("Number is negative :",number)
```
