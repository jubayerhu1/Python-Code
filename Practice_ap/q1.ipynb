## Print numbers from 1 to 5 
### Question write a program to print numbers from 1 to5 using a for loop.
### Expected : output: 1 2 3 4 5
- Hint Use the range () function to loop through number 1 to 5

```python
for i in range(1,6):
    print(i)
```

## 3. Print Even Number from 1 to 10 
- Question : Write a program to print all even numbers from 1 to 10
- Expected Output : 2, 4,6,8,10
- Hit: Check if a number is even using number % 2 == 0.

```python
for i in range(1,11) :
    if i % 2 == 0:
        print(i, end = " ")
```

## 4. Calculate Sum of Number from 1 to 10 
- Question : Write a program to calculate the sum of numbers from 1 to 10.
- Expected Output : sum = 55
- Hint : Initilize a variable for the sum and add each number in the loop


```python
total = 0

for i in range(1,11):
    total += i
print(f" Sum is {total}")
```
