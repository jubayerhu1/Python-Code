## Print numbers from 1 to 5 
### Question write a program to print numbers from 1 to5 using a for loop.
### Expected : output: 1 2 3 4 5
- Hint Use the range () function to loop through number 1 to 5

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,6):
    print(i, end= " ")
```

```python
### 2. Print Squares of Numbers from 1 to 5 
- Question : Write a program to print the squares of numbers from 1 to 5
- Expected Output : 1 4 9 16 25
- Hint : Use number  ** 2 to get the square of each number in the loop
```

```python
for i in range(1,6):
    print(i**2, end = " ")
```
