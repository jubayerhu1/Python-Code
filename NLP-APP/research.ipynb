# Api key set  with gemmin ai

```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY") # it come for env file 
print(GEMINI_API_KEY)
```

```python
user_text1 = """give me the sentiment of this sentence: 
    "I hate this movie"""
user_text2 ="""Give me hindi transilation of this sentence :
        "i hate this movie" 
        """
user_text3 ="""Give me Bangla transilation of this sentence :
        "i hate this movie" one line 
        """
user_text4 = """ Detect the language of this sentence 
    "i hate this movie"

    """
```

```python

genai.configure(api_key = GEMINI_API_KEY)
#model = genai.GenerativeModel("gemini-2.5-flash")
model = genai.GenerativeModel("gemini-2.5-flash")
res = model.generate_content(user_text1)
result =res.text
print(result)
```
