# Exception 
- if things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened
- Exception are rised by python runtime (runtime error)
- You have to takle is on the fly

#### Examples:
- Memory overflow
- Divide by - -> logical error
- Database error

```python
def divided(num):
    result = 12/num
    print(result)

divided(2) #ok to divided 
```

```python
divided(0) # error of ZeroDivisionError: division by zero
# program is fine, when run it will show error it is call run time error.
```

```python
# Let's create a file | write mode 

with open('files/sample.txt', 'w') as f:
    f.write("Hellow world")
```

```python
# Let's read  a file | read mode 
with open('files/sample4.txt', 'r') as f: # [Errno 2] No such file or directory: 'files/sample4.txt'
    read = f.read()
    print(read)
```

```python
# Now use by try and except:
try:
    with open("files/sample4.txt", 'r') as f:  # Here is not correct file , code is good , is not showing error . its show custome message.
        read = f.read()
        print(read)
except:
    print("File not found")
```
