# Loop through a Dictionary
## you can loop through a dictionary by using a -- for -- loop
 - when looping through a dictionary, the return value are the keys of the dictionary, but there are methonds toreturn the values as well.
 

```python
d ={ "name": "jubayer", "age": 33, "gender":"Male"}

for x in d:  # Print all key names in the dictionary , one by one:
    
    print(x)
```

```python
# Print all ---values--- in the dictionary , one by one:
for x in d:
    print(d[x])
```

```python
# you can also use the  - values() method to return values of a dictionary

for x in d.values():  # values(). return all values inside dict
        print(x)
```

```python
# You can use the keys () method tor return the keys of a dictionary:
# keys()-- return all keys 

for x in d.keys():
    print(x)
```

```python
# Loop through both keys and values , by using the -- items()-- 
for x,y in d.items():
    print(x,y)
```
