### remove _ html _tags


```python
import re
 
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
text = "<html><body><p> Movie 1</p><p> Actor - Aamir Khan</p><p> Click here to <a href='http://google.com'>download</a></p></body></html>"
```

```python
remove_html_tags(text)
```

```python
# Apply all columns remove html tags

df['review'] = df['review'].apply(remove_html_tags)
```

### remove _url 

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
text1 = 'Check out my youtube https://www.youtube.com/dswithbappy'
text2 = 'Check out my linkedin https://www.linkedin.com/in/boktiarahmed73/'
text3 = 'Google search here www.google.com'
text4 = 'For data click https://www.kaggle.com/'
```

```python
remove_url(text1)
```

```python

print(remove_url(text2))
print(remove_url(text3))
print(remove_url(text4))
```

## Punctuation handling

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
exclude 
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
text = 'string. with. punctuation ?'
```

```python
start= time.time()
print(remove_punc(text))
time1 = time.time() - start
print(time1 * 50000)
```

```python
# 2nd way to punchation remove punctuation 

def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
start = time.time()
remove_punc1(text)
time2 = time.time() - start
print(time2 * 5000)
```

## Char_conversion handle 


```python
chat_words = {
    'AFAIK' : 'AS Far As I know',
    'AFK'   : 'Away from keyboard',
    'ASAP'  : 'AS soon as possible'
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)

    return " ".join(new_text)
```

```python
chat_conversion('Do this work ASAP')
```
