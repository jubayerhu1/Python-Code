```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# prepare Data (Actul vs predicted)

y_actual = np.array([100, 90, 80, 70])

y_pred = np.array([98, 92, 78, 75])
```

```python
# canculate Absolute and squared metrics

mae = mean_absolute_error (y_actual, y_pred)

mse = mean_squared_error(y_actual, y_pred)

rmse = np.sqrt(mse)  # it has numpy
```

```python
rmse = mean_squared_error(y_actual, y_pred)
rmse
```

```python
print(f'MAE :  {mae}')
print(f'MSE :  {mse}')
print(f'RMSE : {rmse}')
```

```python
# calculate R2 score

r2 = r2_score(y_actual, y_pred)

print(f'R2 score : {r2}')
```

```python
# Adjeste | calculate r2 ( manul formula )

n = len(y_actual)
p =1

adj_r2 = 1-(1-r2) * (n-1)/(n-p-1)

print(f'Adjusted r2 : {adj_r2}')
```
