```python
# 3. create Simple calculator in Python

# To make this work , we use input () to get data from the user and if statement 
# to decide which math operation to perform.

# 1 Get the first number 
num1 = float(input("Enter first number :  "))

# 2. Get the operator the ( the aromatic sign)
op = input("Enter operator ( +,-,*,/) : ")

# 3. Get the scond number 
num2 = float(input("Enter second number : "))

# Logic to calculate the result 
if op == "+":
    print("Total : ", num1+num2)
elif op == "-" :
    print("Total : ", num1-num2)
elif op == "*" :
    print("Total ", num1 * num2)
elif op == "/" :
    if num2 !=0 :
        print("Total :", num1/num2)
    else:
        print("Error : Cannot divide by zero ! ")
else:
    print(" Invalid operator !")
```

```python
# 3 Factorial of a number 
# A factorial is just a multiplication countdown. You take a number and multiply it by every whole number below it until you hit 1.

def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
    
number = int(input("Enter your number : "))
print("Factorial : ",factorial(number))
```
