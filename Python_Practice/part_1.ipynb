## 1 Coverting an Integer into Decimals

```python
## 1 Coverting an Integer into Decimals
import decimal
integer = 10
print(decimal.Decimal(integer))
print(type(decimal.Decimal(integer)))
```

## 2 Converting an String of integers into Decimals

```python
import decimal
string = '12345'
print(decimal.Decimal(string))
print(type(decimal.Decimal(string)))
```

## 3 Reversing a String using an Extended slicing Technique

```python
string = "Python programming"
print(string[:])
print(string[::])
print(string[::-1])
```

## 4 Counting Vowels in a given word

```python
## 4 Counting Vowels in a given word
vowel = ['a', 'e','i', '0','u']
word = 'programming'
count = 0

for character in word:
    if character in vowel:
        count += 1
print(count)
```

## 5 Counting Consonants in a Given word

```python
vowel = ['a', 'e', 'o', 'i','u']
word = 'programming'
count = 0

for character in word :
    if character not in vowel:
        count += 1
print(count)
```

## 6 Counting the number of occurances of character in a string 


```python
## 6 Counting the number of occurances of character in a string 
word = "programming"
character = "g"

count = 0
for i in word :
    if i == character:
        count +=1
print(f"Total occurance : {count}")
```

## 7 writing FIBONACCI Series

```python
# writing FIBONACCI Series
fib =[0,1]

# Range starts from 0 by default

n = 5 

for i in range (n):
    fib.append(fib[-1] + fib[-2])
# Converting the list of integers to string
print(' , '.join(str(e) for e in fib))
```
