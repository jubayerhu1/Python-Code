# # Pandas Series 
### what is a series 

- A pandas series is like a column in a table 
- It is a one dimensional array holding data of any type.

```python
# pandas series 
a = [1,7,2]
my_var = pd.Series(a)
print(my_var)
```

### Label 
- If nothing else is specified, the values are labeled with their index number. First value has index 0,second value has index 1 etc
- This label can be used to access a specified value



```python
print(my_var[0])
print(my_var[2])
```

```python
# Create labels 
# with the index argument , you can name your own labels.
a = [1,7,2]
my_var = pd.Series(a, index = ["x", "y","z"])
print(my_var)

my_var1 = pd.Series([1,3,4], index = ["juba","hus","jih"])
print(my_var1)
```

```python
# When you have crated labels , you can access an item by refferring to the label.

print(my_var1["juba"])
```

## DataFrames
### What is the dataframe?
- A pandas DataFrame is a 2 dimensional data structure , like a 2 dimensional array, or a table with rows and columns.

```python
import pandas as df 
data = {
    "calories" : [420,380,396],
    "duration" : [50,60,45]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Locate Row 
# As you can see from the result above, the DataFrame is a like a table with rows and columns.
# pandas use the -- loc - - attribute to return one or more specific row(s)

# refer to the row index

print(df.loc[0])
```

```python
# Return row 0 and 1 :
print(df.loc[[0,1]])
```

```python
# Name Indexes 

# Add a list of names to give each row a name 

data = {
    "calories" : [420,380,390],
    "duration" : [50,40,45]
}

df = pd.DataFrame(data, index= ["day1","day2","day3"])
print(df)
```

```python
# Locate Named Indexes 
# use the named index in -- loc -- attribute to return the specific row(s)

# Return " day2"

print(df.loc["day2"])
```
