```python
## Sort the data  ascending = True or decending 
coffee.head()
```

```python
coffee.sort_values(["Units Sold"], ascending = True) # ascending means from small to big
coffee.sort_values(["Units Sold"], ascending = False) # Deascending means from big to small
```

```python
# Adding two column for ascending and decending 
coffee.sort_values(["Units Sold"], ascending = True) # ascending means from small to big
coffee.sort_values(["Units Sold","Coffee Type"], ascending = [1,0]) # Deascending means from big to small
```
