```python
import pandas as pd     
import numpy as np
import statistics as st
```

```python
# =========================================
"""
IQR 
1. Sort the data
2. Calculate Q1(25%), and Q3(75%)
3. IQR = Q3 - Q1
4. Find the lower Fence (Q1 - 1.5*(IQR))
5. Find the Uper fence (Q3 + 1.5*(IQR))





"""
# ===========================================
```

```python
data = [1,2,2,2,3,3,4,5,5,5,6,6,6,7,8,9,27,100,102]
data = sorted(data)
print(data)
```

```python
Q1, Q3 = np.percentile(data, [25,75])
```

```python
print(Q1 , Q3)
```

```python
# IQR = Q3 - Q1
IQR = (Q3 - Q1)
print(IQR)
```

```python
Lower_Fence = Q1 -1.5*(IQR)
Lower_Fence
```

```python
Upper_Fence = Q3 + 1.5*(IQR)
Upper_Fence
```

```python
# Non outlier 
non_outlier = []

for x in data:
    if x > Lower_Fence and x < Upper_Fence:
        non_outlier.append(x)
print(x)
```
