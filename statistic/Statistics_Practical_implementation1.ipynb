```python
#  Measure of Central Tendency 
"""
- 1. Mean
- 2. Median
- 3. Mode

"""

```

```python
ages = [23,24,32,45,12,43,67,45,32,56,32,420]
```

```python
import numpy as np
```

```python
print(np.mean(ages))
```

```python
print(np.median(ages))
```

```python
import statistics as st
```

```python
# mean, median, mode 

print(st.mean(ages))

print(st.median(ages))

print(st.mode(ages))
```

```python
import seaborn as sns
```

```python
# out laier
sns.boxplot(ages)
```

```python
# =============
# 5 Number Summary
"""
IQR

1. sorted data
2. Calculate Q1(25%) and Q3(75%)
3. IQR = Q3 - Q1 
4. Find the lower_fence (Q1 - 1.5*(IQR))
5. Find the Upper _fence (Q3 + 1.5 *(IQR))


"""


```

```python
Q1,Q3 = np.percentile(ages,[25,75])

print(Q1,Q3)
```

```python
IQR = Q3 -Q1
Lower_Fence = Q1 - 1.5*(IQR)
Upper_Fence = Q3 + 1.5*(IQR)

print(Lower_Fence , Upper_Fence )
```

```python
# Measure of Dispersion
'''
1. Varience 
2. Standard Deviation

'''
```

```python
ages = [23,24,32,45,12,43,67,45,32,56,32]

#. Variance 

st.variance(ages)
```

```python
# Standard devation

st.stdev(ages)
```

```python
# Pvariance

st.pvariance(ages)
```

```python
# Math
import math
math.sqrt(st.pvariance(ages))
```
