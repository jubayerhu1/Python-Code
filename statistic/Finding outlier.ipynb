IQR 

1. Sort the data
2. calculate q1(25%), and  q3 (75%)
3. IQR (q3-q1)
4. Find the lower Fence( q1 -1.5(IQR))
5. Find the uper Fence ( q3+1.5(IQR))

```python
d = [1,2,2,2,3,3,4,5,5,5,6,6,6,7,8,9,27]
d = sorted(d)

q1,q3 =np.percentile(d,[25,75])
print(q1,q3)
```

```python
#d = [1,2,2,2,3,3,4,5,5,5,6,6,6,7,8,9,27]
#detect_outliers(d)
```

```python
IQR = q3 - q1
print(IQR)
```

```python
Lower_fence  = q1 -1.5*(IQR)
print(Lower_fence)
```

```python
Uper_Fence = q3 + 1.5*(IQR)
print(Uper_Fence)
```

```python
print(Lower_fence, Uper_Fence)
```
