# Matplotlib

What is Matplotlib?

Matplotlib is a powerful data visulization library in Python for creating, static, animated, and interative plots.
It provides an easy way to generate various types of graphs. it an essential tool for data analysis, machine learning, and scientific computing.

It is widely used in data science, statistic, and AI to visualize trends, patterns, and distribution in data.

```python
"""Bar chart
A bar chart is a type of chart that uses bars to represent data. The length of each bar corresponds to the value of the variable it represents. It is useful for comparing different categories.

Bar charts are used to visually compare categories of data.
"""
blood_group = 'B+', 'B-','AB-'

```

```python
Total_person = [15,25,8]
```

```python
plt.bar(blood_group, Total_person)
```

```python
# Change color 
plt.bar(blood_group, Total_person, color= 'green')
```

```python
# Horizontal Bar 

plt.barh(blood_group,Total_person, color = 'green')

plt.title( 'Blood Group Campaign ')

plt.ylabel("Blood Gropu")
plt.xlabel("Frequecy")

plt.show()
```
