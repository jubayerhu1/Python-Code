Assignment 1: Temperature Check
Problem:
Write a program that takes the current temperature as input and prints:
● "It's too cold!" if below 10°C
● "It's cool outside" if between 10°C and 25°C
● "It's hot outside!" if above 25°C

```python
print("---start----")
tem = int(input("Enter Temperature  : "))



if tem < 10:
    print("It's cold !")
elif tem >10 and tem <= 25:
    print("It's cool outside !")
else:
    print("It's hot outside ")  # tem>25 

print("..end ..")
```

Assignment 2: ATM Withdrawal System
Problem:
Ask the user for their account balance and the amount they want to withdraw.
● If the withdrawal amount is greater than the balance → print "Insufficient
Balance"
● If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.

```python
balance = float(input("Enter your account balance: $"))
withdraw_amount = float(input("Enter the amount you want to withdraw: $"))

if withdraw_amount > balance:
    print("Insufficient Balance")
elif withdraw_amount < 0:
    print("Withdrawal amount cannot be negative.")
else:
    remaining_balance = balance - withdraw_amount
    print("Transaction Successful")
    print(f"Remaining balance: ${remaining_balance:.2f}")
```

Assignment 3: Student Grade Calculator
Problem:
Input marks (0–100) and print grade based on:
● 90–100 → A+
● 80–89 → A
● 70–79 → B
● 60–69 → C
● Below 60 → Fail

```python
print( "Student Grade Calculator  start ....")

marks = float(input("Enter your mark (0-100) : "))

if marks <0 and marks >100:
    print ("Invalid input . Masrks must be between 0 - 100. ")
elif marks >=90 :
    print( "Grade : A+")
elif marks >= 80:
    print("Grade : A")
elif marks >= 70:
    print("Grade : B")
elif marks >= 60:
    print("Grade : C")
else:
    print("Grade : Fail")
```

Assignment 4: Bus Ticket Discount
Problem:
Ask the user for age and print ticket price:
● Below 5 → Free
● 5–18 → 50% Discount
● 60 or above → 30% Discount
● Others → Full Price

```python
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age. Age cannot be negative.")
elif age < 5:
    print("Ticket Price: Free")
elif age >= 5 and age <= 18:
    print("Ticket Price: 50% Discount")
elif age >= 60:
    print("Ticket Price: 30% Discount")
else:
    print("Ticket Price: Full Price")
```
