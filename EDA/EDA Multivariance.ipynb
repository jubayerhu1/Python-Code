## Bar Plot (Numerical to Categorical) we use bar plot when two type data(like numberical and categorical)

```python
titanic.head(1)
```

```python
#sns.barplot(data=df, x="island", y="body_mass_g")
#sns.barplot(data= titanic, x = 'Pclass', y = 'Age')
sns.barplot(data= titanic, x = titanic["Pclass"], y=titanic["Age"])
```

```python
# Multivarient --> sex column. (confidence inter bal think line) 30:
sns.barplot(data= titanic, x = titanic["Pclass"], y=titanic["Age"],hue=titanic["Sex"])
```

```python
# you can use boxplot # it has outlier
sns.boxplot(data= titanic, x ="Pclass", y = "Age")
```

```python
# you can use boxplot # it has outlier
sns.boxplot(data= titanic, x ="Pclass", y = "Age", hue= "Survived")
```

```python
# Displot (Numerical to categorical )
# Dath 
sns.displot(titanic[titanic['Survived'] == 0]['Age'], hist= True)
sns.displot(titanic[titanic['Survived'] == 1]['Age'], hist= True)
```
