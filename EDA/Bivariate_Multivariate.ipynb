# Bivariate 

## Scatterplot (numerical - numerical)
### i want tips relation ship to bill ( we use 2 columns that means Bivariate )
### use seaborn for scatterplot

```python
tips.head()
```

```python
sns.scatterplot(tips['total_bill'])
```

```python
sns.scatterplot(data = tips, x ="total_bill", y ='tip')


plt.xlabel('Total bill')

plt.ylabel('Tips')

plt.title("Tips data")
plt.show()
```

```python
# It is Bivariate 
```

# Multivariate Analysis
## I want add more columns [ hue = 'sex]

```python
sns.scatterplot(data = tips, x ="total_bill", y ='tip', hue = 'sex')


plt.xlabel('Total bill')

plt.ylabel('Tips')

plt.title("Tips data")
plt.show()
```

## I want to separate smoker and non smoker

```python
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue= 'sex', style = tips['smoker'])
plt.xlabel('Total bill')
plt.ylabel('Tips')

plt.title('Tip data')
plt.show()
```

# Outlier
## you can use box plot for outlier

```python
sns.boxplot(data=ti, x = 'Pclass', y = 'Age')

plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title("Titanic")

plt.show()
```

```python
## more than two columns 
sns.boxplot(data=ti, x = 'Pclass', y = 'Age', hue = 'Survived')

plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title("Titanic")

plt.show()
```
