# Editing key value pair

```python
# Adding key -value pair
d1 = {"name" : "Jubayer", "age" : 43, "gender": "male"}
d1["lastName"] = "Hussain"
print(d1) # last name is added your d1 list 

d1["name"] ="Jubayer Hussain"
print(d1) # last one is editing the name of jubayer hussain . but when you try to editing the key modify , it will through error.
```

# Remove key -value pair


```python

# pop() pop alway take input . input means it will key name | sepcific item
d1 = {"name" : "Jubayer", "age" : 43, "gender": "male"}
print(d1)
d1.pop("age")
print (d1)
```

```python
# popitem() it will remove the last one 
d1 = {"name" : "Jubayer", "age" : 43, "gender": "male"}
print(d1)
d1.popitem()# gender key will remove
print(d1) 
```

```python
# del 
d1 = {"name" : "Jubayer", "age" : 43, "gender": "male"}
print(d1)
del d1["name"] # name is deleted 
print(d1)
```

```python
# clear is clear everything on the dictionary

d1 = {"name" : "Jubayer", "age" : 43, "gender": "male"}
print(d1)

d1.clear()
print(d1) # return empty dictionary
```
