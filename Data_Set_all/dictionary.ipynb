```python
# 2d dictionary

d3 = {
    "name": "jubayer",
    "collage" : "ABC",
    "Semester" : 5,
    "Subject" : {"math": 50, "englsih": 45, "phy": 60}


}
print(d3)
print(d3["Subject"]["math"])
```

```python
# using sequence and dic function
d4 = dict([("name","bappy")])
print(d4)
d5 = dict([("name","bappy"),("age",33)])
print(d5)
```
